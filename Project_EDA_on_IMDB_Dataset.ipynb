##Converting the String values to the proper list(Cast, Crew, Genres)
Only for non-null strings



```python
def convert_to_list(str):
  return eval(str)
```

```python
#Applying convert_to_list function on the genres column
imdb_df.loc[~imdb_df['genres'].isna(),'genres'] =imdb_df.loc[~imdb_df['genres'].isna(),'genres'].apply(convert_to_list)
```

```python
#Applying convert_to_list function on the cast column
imdb_df.loc[~imdb_df['cast'].isna(),'cast'] =imdb_df.loc[~imdb_df['cast'].isna(),'cast'].apply(convert_to_list)
```

##4.Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
#For the question we will only consider the rows in the columns having non null values
no_na_crew = imdb_df_new[~imdb_df_new['crew'].isna()]
no_na_crew.shape
```

```python
#Extracting the list of all the producers for the given movie index
def create_producer_list(index):
  movie_index = no_na_crew.iloc[index]
  crew_list = movie_index.loc['crew']
  producer_list = []
  for elem in crew_list:
    if elem['job']=='Producer':
      producer_list.append(elem['name'])
  return producer_list
```

```python
create_producer_list(61)
```

```python
#Extracting the list of all the directors for the given movie index
def create_director_list(index):
  movie_index = no_na_crew.iloc[index]
  crew_list = movie_index.loc['crew']
  director_list = []
  for elem in crew_list:
    if elem['job']=='Director':
      director_list.append(elem['name'])
  return director_list
```

```python
create_director_list(61)
```

```python
#Creating the empty table to store values in it for later
Table = pd.DataFrame(columns=['Movie Title','Producer','Director','ROI'])
```

##Appending the Table Dataframe and also using Try Except block to buypass the errors in the dictonaries containing float as value

```python
for index,row in no_na_crew.iterrows():
  try:
    Table = Table.append({'Movie Title':no_na_crew.loc[index,'original_title'],'Producers':create_producer_list(index),'Director':create_director_list(index),'ROI':no_na_crew.loc[index,'roi']},ignore_index = True)
  except:
      continue
```
